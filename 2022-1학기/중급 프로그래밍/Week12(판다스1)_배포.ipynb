## 데이터 프레임 생성


```python
# 넘파이로 데이터프레임생성
import numpy as np
arr=np.arange(12).reshape(3,4)
arr
df=pd.DataFrame(arr)
df
```

```python
# 딕셔너리로 데이터프레임 생성
name=['장은실','오경선','양숙희']
dept=['컴공','영어','수학']
score=[90,80,95]
df = pd.DataFrame({'이름':name,
                   '점수':score,
                   '전공':dept})
print(df)
```
