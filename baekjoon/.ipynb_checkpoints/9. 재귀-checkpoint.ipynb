### 팩토리얼

```python
# 모듈 사용의 편리함 ㅎㅎ
import math

n = int(input())

math.factorial(n)
```

```python
# 재귀함수로 풀기
def factorial(n):
    if n<= 1:
        return 1
    
    return n * factorial(n-1)

n = int(input())

print(factorial(n))
```

### 피보나치 수 5

```python
n = int(input())

def fivo(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else :
        result = fivo(n-1) + fivo(n-2)
        n -= 1
        fivo(n)
        return result
    
print(fivo(n))
```
