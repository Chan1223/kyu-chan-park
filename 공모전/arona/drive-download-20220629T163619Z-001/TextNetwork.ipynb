```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
#https://gist.githubusercontent.com/chulgil/d10b18575a73778da4bc83853385465c/raw/a1a451421097fa9a93179cb1f1f0dc392f1f9da9/stopwords.txt
with open('stopwords_korean.txt',encoding='cp949') as f:
    stopwords = f.readlines()
    stopwords = [word.strip() for word in stopwords]
```

```python
stopwords.extend(['요','저','제','좀','것','더','군','혹시','보고','거','민','가요','중'])
```

```python
filtered_sent = []
for i, document in tqdm(enumerate(comments)):
    clean_words = []
    for word in okt.pos(document, stem=True): #어간 추출
        if word[1] == 'Noun' and word[0] not in stopwords: #명사, 동사, 형용사
            clean_words.append(word[0])
    document = ' '.join(clean_words)
    filtered_sent.append(document)
```

```python
def do_sna_ko(filtered_content, selected_words):
    g = nx.Graph()
    g.add_nodes_from(selected_words)

    for pair in list(itertools.combinations(list(selected_words),2)):
        if pair[0] == pair[1]:
            continue
        for sent in filtered_content:
            if pair[0] in sent and pair[1] in sent:
                if pair in list(g.edges()) or (pair[1],pair[0]) in list(g.edges()):
                    g[pair[0]][pair[1]]['weight'] += 1
                else:
                    g.add_edge(pair[0],pair[1],weight=1)

    
    pos = nx.shell_layout(g)
    nx.draw(g,pos,font_family="Malgun Gothic")
    nx.draw_networkx_labels(g,pos,font_family="Malgun Gothic",font_size=25)
    plt.show()
    return g
```
