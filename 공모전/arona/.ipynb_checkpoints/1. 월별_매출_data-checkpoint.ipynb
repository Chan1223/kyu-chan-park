```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/chan/Documents/공모전/월별_매출_data.csv')
```

```python
# '아로나민'이 들어간 제품만 추출
df1 = df[df['제품'].str.contains('아로나민')]
```

```python
# 날짜별로 재정렬
df1 = df1.sort_values(['제품', '년'])
```

```python
df1
```

```python
# ATC. 업체명, 제조사 제거 및 index 초기화
df2 = df1[['년','월', '제품','판매금액', '판매수량']]
df2 = df2.reset_index()
```

```python
df2 = df2[['년','월', '제품','판매금액', '판매수량']]
df2
```

```python
df2['제품'].unique()
```

```python
# 년 / 월 합치기
df2['month'] = df2['년'] + " " +df2['월']
```

```python
df2
```

```python
df2 = df2[['month', '제품', '판매금액' , '판매수량']]
df2
```

```python
# 아로나민 data만 우선 확인
df3 = df2.loc[df2['제품'].str.contains('아로나민골드정')]
```

```python
df3
```

```python
df3['month'] = df3['month'].replace('년', ' ')
```

```python
df3
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
x = df3['month']
y = df3['판매금액']


plt.figure(figsize= (50, 50))
plt.xlabel('month', fontsize=20)
plt.ylabel('sales', fontsize=20)
# plt.axvline(x='2017', color='k', linestyle='--', linewidth=1)
# plt.axvline(x='2018', color='k', linestyle='--', linewidth=1)
# plt.axvline(x='2019', color='k', linestyle='--', linewidth=1)
# plt.axvline(x='2020', color='k', linestyle='--', linewidth=1)
# plt.axvline(x='2021', color='k', linestyle='--', linewidth=1)
plt.axvline(x='2022', color='k', linestyle='--', linewidth=1)
plt.plot(x, y)
```
