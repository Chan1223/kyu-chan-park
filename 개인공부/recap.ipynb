```python
import numpy as np
import matplotlib.pyplot as plt
```

# MSE

```python
# 간단한 linear 함수
def linear(x):
    return x*1.2
```

```python
# test data
x = np.arange(-5, 5, 0.1)

x = np.random.randn(100, 1)
x = np.array(sorted(x))
y = linear(x)
```

```python
plt.plot(x, y)
plt.plot([0,0],[10.0,-20.0], ':') # 가운데 점선 추가
plt.title("Linear")
plt.show()
```

```python
# mse graph 
plt.plot(x, (y-x)**2)
```

# Sigmoid 함수

```python
# logistic function
def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)
```

```python
# sigmoid 하나만 그렸을 때
plt.plot(x, y)
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title("Sigmoid Function")
plt.show()
```

```python
# w 값의 변화에 따른 sigmoid 함수의 변화
x = np.arange(-5, 5, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(1*x)
y3 = sigmoid(2*x)


plt.plot(x, y1, "r", linestyle='--')
plt.plot(x, y2, "g")
plt.plot(x, y3, "b", linestyle='--')

plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title("Sigmoid Function")
plt.show()
```

```python
# bias 값의 변화에 따른 sigmoid 함수의 변화
x = np.arange(-5, 5, 0.1)
y1 = sigmoid(x-1)
y2 = sigmoid(x) 
y3 = sigmoid(x+1)


plt.plot(x, y1, "r", linestyle='--')
plt.plot(x, y2, "g")
plt.plot(x, y3, "b", linestyle='--')

plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title("Sigmoid Function")

plt.show()
```

```python
# 미분식
def diff_x(a, h = 1e-5):
    return (sigmoid(a + h) - sigmoid(a)) / h
```

```python
# bias 값의 변화에 따른 sigmoid 함수의 변화
x = np.arange(-5, 5, 0.1)
y1 = diff_x(x-1)
y2 = diff_x(x) 
y3 = diff_x(x+1)


plt.plot(x, y1, "r", linestyle='--')
plt.plot(x, y2, "g")
plt.plot(x, y3, "b", linestyle='--')

plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title("Sigmoid Function")

plt.show()
```

# 로지스틱에 MSE를 사용한다면

```python
# test data, 100점 만점 중 80정 이상만 합격(1), 아니면 불합격(0)
x = np.random.randn(10000, 1)*10
x = np.array(sorted(x))
y = []
for i in x:
    if i >0.2:
        y.append(1)
    else :
        y.append(0)
        
```

```python
x = x.reshape(10000,1)
y = np.array(y).reshape(10000,1)
```

```python
# y값 확인
plt.scatter(np.arange(0, 1 , 0.0001), y) 
plt.show()
```

```python
# 예측 안한 확률값 상태에서 그래프
plt.plot(y-(sigmoid(x))**2)
```

```python
# 분류 문제에서의 MSE를 사용하지 못하는 이유, non-convex 곡선이 나옴
plt.plot(y- (sigmoid(x)>0.3)**2)
```

# Cross-entropy

```python
# 예측값이 1일 경우
# 예측값이 1에 가까울수록 cost가 낮아짐 
plt.plot(np.arange(0, 1 , 0.0001), -np.log(sigmoid(x)) ,'b', linestyle='--')
```

```python
# 예측값이 0일 경우
# 예측값이 0에 가까울수록 cost가 낮아짐 
plt.plot(np.arange(0, 1 , 0.0001), -np.log((1-sigmoid(x))) ,'b', linestyle='--')
```

```python
# 2가지를 혼합한 분류 문제에 사용할 수 있는 새로운 cose function
plt.plot(np.arange(0, 1 , 0.0001), -np.log(sigmoid(x)) ,'r', linestyle='--')
plt.plot(np.arange(0, 1 , 0.0001), -np.log((1-sigmoid(x))) ,'b', linestyle='--')
plt.show()
```

# softmax

```python
x = np.random.randn(1,5)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    exp_sum = np.exp(x).sum()
    return  exp_x / exp_sum
```
