* 자연수(natural number) : 1, 2, 3, 4, 5와 같은 수로 양수만 포함됨

* 범자연수(whole number) : 자연수에 '0'이란 개념이 받아들여진 체계
* 정수(integer) : 0 뿐만 아니라 양수와 음수의 자연수가 포함되는 개념
* 유리수(rational number) : 2/3 과 같이 분수로 표현할 수 있는 숫자, 이산적인 단위로 측정할 수 없는 수량으로 인해 필요함
* 무리수(irrational number) : 파이, 제곱근, 오일러 수 e 등이 이에 해당
* 실수(real number) : 유리수와 무리수를 포함하는 개념
* 복소수와 허수(complex number, imaginary number) : 실제로 존재하지 않는 허수와, 실수와 허수의 합으로 나타낸 복소수

### 합계(summation)
* 합계 기호 $\Sigma$(sigma) 는 여러개를 더한다는 의미

```python
# generator를 이용한 합 연산
summation = sum(2*i for i in range(1,6))
print(summation)
```

```python
# 네 배열의 합 구하기
x = [1, 4, 6, 2]
n = len(x)

summation = sum(10*x[i] for i in range(0, n))
print(summation)
```

```python
# sympy를 이용한 연산, lazy operation이기 때문에 doit 없이는 실행되지 않고 수식을 보여줌
from sympy import *

i, n = symbols('i n')

# 1부터 n까지 i를 반복하며 2를 곱한 뒤 덧셈
summation = Sum(2*i, (i, 1, n))
summation
```

```python
# n을 5로 지정
up_to_5 = summation.subs(n, 5)
up_to_5
```

```python
up_to_5.doit() # 결과 출력시 30이 나옴, lazy evaluation이기 때문에 doit시 실행됨
```

### 미분

```python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m

def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function, 2, 1e-9)

print(slope_at_2)
```
