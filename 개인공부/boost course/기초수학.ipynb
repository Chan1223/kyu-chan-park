```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

### 일차함수

그래프 상에서 직선인 그래프(linear)

$y=ax+b$  
- $a$ : 기울기  
- $b$ : $y$절편

```python
def linear_function(x):
    a = 0.5
    b = 2
    return a*x + b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5, 5, 0.1)
y = linear_function(x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear function')
plt.show()
```

### 이차함수
일반적으로 두개의 실근을 가짐  

$y = ax^2 + bx + c$


```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    return a*x**2 + b*x + c
```

```python
x = np.arange(-5, 5, 0.1)
y = quadratic_function(x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('quadratic_function')
plt.show()
```

### 삼차함수
$ y = ax^3 + bx^2 + cx + d $ 

```python
def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    return a*x**3 + b*x**2 + c*x + d
```

```python
cubic_function(3)
```

```python
x = np.arange(-5, 5, 0.1)
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('cubic function')
plt.show()
```

### 함수의 최소값/최대값

```python
def my_func(x):
    a = 1
    b = -3
    c = 10
    return a*x**2 + b*x + c
```

```python
x = np.arange(-10, 10, 0.1)
y = my_func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(1.5, my_func(1.5))
plt.text(1.5-1.5, my_func(1.5)+5, f'min value of f(X)\n({1.5}. {my_func(1.5)})', fontsize=10)
plt.title('my_func')
plt.show()
```

### 특정 구간 내에서 최소값 구하기

```python
def get_minimum(x1, x2, f):
    x = np.arange(x1, x2, 0.01)
    y = f(x)
    
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('get_minimum')
    plt.show()
    return min(y)
```

```python
get_minimum(1, 4, my_func)
```

### 지수함수 / 로그함수
* 지수함수-로그함수는 역함수 관계, $y=x$ 인 직선 대칭 (단, 밑이 같을 때)
* 파이썬으로 직접 구현 가능


### 지수함수
* $y = a^x(a\ne0)$(기본형)
* $y = e^x (e=2.71828...)$

```python
def exponential_function(x):
    a = 4
    return a**x
```

```python
def exponential_function2(x):
    a = 4
    return math.pow(a, x)
```

```python
x = np.arange(-3, 2, 0.1)
y = exponential_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 15)
plt.xlim(-4, 3)
plt.title('exponential_function')
plt.show()
```

### 역함수 관계
* $ y=x$ 대칭

```python
x = np.arange(-1, 5 ,0.01)
y1 = np.exp(x)

x2 = np.arange(0.000001, 5, 0.01)
y2 = np.log(x2)

y3 = x

plt.plot(x, y1, 'r-', x2, y2, 'b-', x, y3, 'k--')

plt.ylim(-2, 6)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 함수 조작 
* $y = -\log_a(x)$ 와 $y = -\log_a(1-x)$  
$x=0.5$ 대칭  
* Logistic Regerssion을 위한 함수


```python
x = np.arange(-10, 10, 0.01)
y1 = -np.log(x)
y2 = -np.log(1-x)

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')

plt.grid()
plt.plot(x, y1, 'b-', x, y2, 'r-')
plt.text(0.9, 2.0, 'y=-log(1-x)', fontsize=15)
plt.text(0.1, 3, 'y=-log(x)', fontsize=15)

plt.xlim(-0.3, 1.4)
plt.ylim(-0.5, 4)

plt.scatter(0.5, -np.log(0.5))
```
