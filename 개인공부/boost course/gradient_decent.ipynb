```python
import numpy as np
import sympy as sym
from sympy.abc import x # 사용할 변수명 불러오기
import matplotlib.pyplot as plt
```

# sympy 활용

```python
# 함수가 다른 변수로 치환될 수 있게끔 만들어주는 함수
# 치환된 함수와, 치환되기 전 함수를 반환함
def func(val):
    fun = sym.poly(x**2 + 2*x + 3)
    return fun.subs(x, val), fun
```

```python
# 함수 미분
# 미분된 함수값과 치환된 미분값을 반환함
def func_gradient(fun, val):
	_, function = fun(val)
	diff = sym.diff(function, x)
	return diff.subs(x, val), diff
```

```python
# 경사하강법
def gradient_descent(fun, init_point, lr_rate=1e-2, epsilon=1e-5):
    cnt = 0
    val = init_point # 초기값
    diff, _= func_gradient(fun, val) # 시작위치에서의 첫 gradient
    while np.abs(diff) > epsilon: # 최소 점이라고 생각할 수 있는 epsilon보다 클 경우 계속 반복
        val -= lr_rate * diff # 미분값에 학습률을 곱하여 위치를 갱신
        diff, _ = func_gradient(fun, val) # 다시 경사도 계산
        cnt += 1 # 수행한 횟수 count

    print("함수: {}\n연산횟수: {}\n최소점: ({}, {})".format(fun(val)[1], cnt, val, fun(val)[0]))
```

```python
gradient_descent(fun=func, init_point=np.random.uniform(-2,2))
```
