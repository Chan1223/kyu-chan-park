#### https://www.acmicpc.net/problem/1012

```python
import sys
from collections import deque

input = sys.stdin.readline

def bfs(arr, x, y, visited, M, N):
    # 특정 포인트 기준으로 동서남북 이동을 위한 d값 형성
    dx = [0, 1, 0, -1]
    dy = [1, 0, -1, 0]

    que = deque()
    que.append((x,y))
    visited[y][x]= 1

    while que:
        cx, cy = que.popleft()

        for i in range(4):
            nx = cx + dx[i]
            ny = cy + dy[i]

            if 0 <= nx < M and 0 <= ny < N:
                if visited[ny][nx] == 0 and arr[ny][nx] == 1:
                    visited[ny][nx] = 1 
                    que.append((nx, ny))
                    



T = int(input())

for i in range(T):  # T 횟수만큼 반복
    M, N, K = map(int, input().split())
    
    arr = [[0] * M for i in range(N)]    # 초기 배추밭 크기 형성
    visited = [[0] * M for i in range(N)]# 방문여부를 확인할 visited 구성

    for _ in range(K):                   # K 줄만큼 배추 input 받기
        x, y = map(int, input().split()) # x, y를 받기 때문에 뒤집어줘야함에 주의
        arr[y][x]=1                      # 뒤집힘에 유의하며 해당 지역에 배추 심기

    cnt = 0
    for y in range(N):                   # 심어진 배추를 따라 bfs 수행
        for x in range(M):
            if visited[y][x] == 0 and arr[y][x]==1:
                bfs(arr, x, y, visited, M, N)
                cnt += 1
    print(cnt)
```
