#### https://www.acmicpc.net/problem/1012

```python
import sys
from collections import deque

input = sys.stdin.readline

def bfs(arr, x, y, visited, M, N):
    #       d 
    dx = [0, 1, 0, -1]
    dy = [1, 0, -1, 0]

    que = deque()
    que.append((x,y))
    visited[y][x]= 1

    while que:
        cx, cy = que.popleft()

        for i in range(4):
            nx = cx + dx[i]
            ny = cy + dy[i]

            if 0 <= nx < M and 0 <= ny < N:
                if visited[ny][nx] == 0 and arr[ny][nx] == 1:
                    visited[ny][nx] = 1 
                    que.append((nx, ny))
                    



T = int(input())

for i in range(T):  # T  
    M, N, K = map(int, input().split())
    
    arr = [[0] * M for i in range(N)]    #    
    visited = [[0] * M for i in range(N)]#   visited 

    for _ in range(K):                   # K   input 
        x, y = map(int, input().split()) # x, y    
        arr[y][x]=1                      #      

    cnt = 0
    for y in range(N):                   #    bfs 
        for x in range(M):
            if visited[y][x] == 0 and arr[y][x]==1:
                bfs(arr, x, y, visited, M, N)
                cnt += 1
    print(cnt)
```
