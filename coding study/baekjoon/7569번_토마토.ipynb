### https://www.acmicpc.net/problem/7569

```python
from collections import deque
import sys

input = sys.stdin.readline

def solution(M, N, H):
    que = deque()
    arr = []

    for h in range(H):
        col = []
        for n in range(N):
            row = list(map(int, input().split()))
            for m in range(M):
                if row[m] == 1:
                    que.append((h, n, m, 0))
            col.append(row)
        arr.append(col)


    dx = [0, 1, 0, -1, 0, 0]
    dy = [1, 0, -1 ,0, 0, 0]
    dz = [0, 0, 0, 0, 1, -1]

    max_day = 0

    while que:
        z, y, x, day = que.popleft()
        max_day = max(max_day, day)
        
        for i in range(6):
            nx = x + dx[i]
            ny = y + dy[i]
            nz = z + dz[i]
            if 0 <= nx < M and 0 <= ny < N and 0 <= nz < H and arr[nz][ny][nx] == 0:
                arr[nz][ny][nx] = 1
                que.append((nz, ny, nx, day+1))

    for h in range(H):
        for n in range(N):
            for m in range(M):
                if arr[h][n][m] == 0:
                    return -1
    return max_day           

M, N, H = map(int, input().split())

print(solution(M, N, H))
```
