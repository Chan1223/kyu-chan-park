```python
count = 0
n = 1 

while n < num :
    for i in range(0, len(str(n))):
        print(str(n)[i])
    n + 1
```

```python
count = 0
n = 1 

while n < num :
    for i in range(0, len(str(n))):
        if n[i] % 3 == 0:
            count += 1
    n + 1
```
