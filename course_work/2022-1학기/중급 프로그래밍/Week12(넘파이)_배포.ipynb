1. 모든 객체(Object) 에게는 저마다의 클래스(Class) 가 존재한다.
2. 메소드(Method) 란, 클래스에서 정의되는 함수로, 객체의 동작을 표현한다.
3. 속성(Attribute) 이란, 클래스에서 정의되는 변수로, 각 객체의 고유값을 나타낸다.
4. 모든 메소드는 반드시 첫번째 인수로 self 를 가집니다.
5. 특별 메소드(Special Method) 란, 파이썬에서 특별하게 취급되는 메소드로 이름이
앞 뒤로 언더바 두개(__)로 구성되어 다른 메소드와 구별된다.
6. 생성자(Constructor) 란, 특별 메소객체를 새로이 생성할 때 사용되는 메소드이며,
객체를 초기화하는 역할을 가지고 있으며 별도의 return 문은 없다. 클래스명(인수1, 인수2, ...)와 같이 호출한다.

- 객체(object) : 속성(특징)과 동작을 가질 수 있는 모든 것
- 객체지향프로그래밍 : 객체를 기반으로 프로그램을 만드는 프로그래밍 언어로 소프트웨어를 만드는 가장 효과적인 방법 중 하나
- 클래스 : 객체의 설계도로 속성+메소드로 구성됨
- 인스턴스(instance) : 클래스를 기반으로 생성한 객체
- 생성자(constructor) : 인스턴스를 생성할 때 사용하는 함수
- 속성(attribute) : 클래스가 가진 변수
- 메소드(method) : 클래스가 가진 함수

클래스를 정의 -> 객체를 생성- > 객체를 이용하여 프로그램 작성

* 클래스 속성
 - 모든 객체에서 접근 가능
 - 모든 객체에서 공유되는 속성을 원한다면 클래스 속성을 사용 
 - 클래스명.속성명
* 인스턴스 속성
 - 해당 객체에서만 접근 가능
 - 해당 객체에서만 그 값이 유효하다. 
 - self.속성명
 
* 상속(inheritance)은 기존에 존 재하는 클래스로부터 코드와 데이터를 이어받고 자신이 필요한 기 능을 추가하는 기법


* 메소드 오버라이드(재정의)
 - 부모 클래스의 메소드는 자식 클래스가 자신의 필요에 맞추어서 변경한다 
 - “자식 클래스의 메소드가 부모 클래스의 메소드를 오버라이드(재정의)한다”고 말한다.

# 넘파이 연산
- 전통적인 배열은 동일한 자료형을 가진 데이터를 연속적으로 저장한다.
- 넘파일 배열은 C언어 기반한 배열 구조로 메모리를 적게 차지하고 속도가 빠르다.
- 넘파이 배열을 사용하면 배열 간에 수학적인 연산을 적용
- 고급 연산자와 풍부한 함수를 제공한다.

```python
mid=[10,20,30]
final=[10,20,30]
```

```python
print(np.add(mid, final))
```

```python
# 리스트
m= len(mid)
total,avg = [],[]

for i in range(m):
    total.append(mid[i]+final[i])
    avg.append(total[i]/2)
print(total)
print(avg)
```

```python
# 넘파이 배열
mid = np.array(mid)
final = np.array(final)
total = mid + final
print(total)
print(total/2)
```

```python
a = np.arange(1,11)
b = np.arange(1,11)
print(a+b)
print(np.add(a,b))
print(a-b)
print(np.subtract(a,b))
print(a*b)
print(np.multiply(a,b))
print(a/b)
print(np.divide(a,b))
```

```python
# 벡터 수행시간
import time

a = np.arange(1000000)
result = 0
s=time.time()
for v in a:
    result += v
e = time.time()
print(result, e-s)

s=time.time()
result=np.sum(a)
e=time.time()
print(result,e-s)
```
