```python
import tensorflow as tf
import pandas as pd
import numpy as np
from transformers import TFBertForNextSentencePrediction
from transformers import AutoTokenizer
```

```python
model = TFBertForNextSentencePrediction.from_pretrained('bert-base-uncased')
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
data_origin = pd.read_csv('su.csv')
data = data_origin.loc[data_origin['type']=='blank']
```

```python
data = data[['question', 'answer', '1', '2', '3', '4', '5']]
data = data.reset_index(drop=True)
```

```python
temp = []
for i in range(len(data)):
    a = data['question'][i].split('_')[0]
    temp.append(a)
```

```python
data['question'][0]
```

```python
temp[0]
```

```python
prepro = pd.DataFrame(temp)
```

```python
data['prepro']=prepro
```

```python
data = data[['prepro', 'answer', '1', '2', '3', '4', '5']]
data
```

```python
# 상관없는 두 개의 문장
total = []
for i in range(len(data)):
    temp = []
    prompt = data['prepro'][i]
    for j in range(1, 6):
        next_sentence = data[str(j)][i]
        encoding = tokenizer(prompt, next_sentence, return_tensors='tf')

        logits = model(encoding['input_ids'], token_type_ids=encoding['token_type_ids'])[0]

        softmax = tf.keras.layers.Softmax()
        probs = softmax(logits)
        # print('최종 예측 레이블 :', tf.math.argmax(probs, axis=-1).numpy())

        temp.append(probs)

    temp = np.array(temp).reshape(5,2)
    temp = temp.reshape(10,1)
    temp = temp[1::2]

    predict_answer = np.argmax(temp)
    total.append(predict_answer)
    print(i)
```

```python
a = np.array(data.answer)
```

```python
b = np.array(total)+1
```

```python
correct_num = (a == b).sum()
```

```python
question_num = len(data)
```

```python
correct_num/question_num
```

```python
data['pred'] = b 
```
